Taking a glance into the dataframe, we can observe 
- some variables are corresponding to each other eg `MakeCode`corresponds to the indexing code for `Make`, so as `FamilyCode` vs. `Model`.
- variable `Description` are additional info about the car which consists of strings which may not be helpful for model building. The columns are information breakdown of `Description` such as `Series`, `SeriesModelYear`, `BadgeDescription`, `BodyStyleDescription`, `BodyConfigDescription` etc. 
- following variables are `wheelBaseCOnfig`, `Roofline`, `ExtraIdentification` which consists of many missing values. 
- followed by variables describing the drive, gear, engine, fuel, cylinders, electric voltage etc. These variables are related to the performance / condition of a car. 
- The following variables are related to the environmental impact of the car. Eg. `GreenHouseRating`, `AirPollutionRating`, CO2 emmisions, `EmissionStandard` etc. 
- `AncapRating` an indicator of car safety by the standards of [The Australasian New Car Assessment Program](https://www.ancap.com.au/safety-ratings-explained)  are provided. 
- Some [VFACT](https://www.carexpert.com.au/category/vfacts) parameters on class, segment, price are available.
